```python
dataset_Date_col = dataset['Date'].unique()  # Guardo en una variable los valores de la columna 'Date', que utilizare para poder generar graficos del conjunto de datos.
# Muestro los posibles valores de cada una de estas caracteristicas.
dataset_Date__count = dataset['Date'].value_counts()

dataset_Date__count
```

```python
# Proceso el array con los registros sin duplicados, y defino un diccionario que contiene el numero de coincidencias con respecto a un mes dado.
fechas_post_corte = dataset_Date_col  # Esta lista ya debería estar filtrada para incluir solo fechas después de la fecha de corte

month_count_hashmap = {}

for fecha in fechas_post_corte:
    elements_day = fecha.split('/')
    mes = elements_day[1]
    anio = elements_day[-1]
    registro_str = f'{mes}_{anio}'

    # Incrementar el contador para el mes y año
    if registro_str in month_count_hashmap:
        month_count_hashmap[registro_str] += 1
    else:
        month_count_hashmap[registro_str] = 1

# Ahora, month_count_hashmap contiene el recuento de fechas para cada combinación mes-año
```

```python
# Es importante recordar que antes de continuar con cualquier tipo de procesamiento, es importante convertir los valores de las columnas no numericas a numericas, a fin de poder operar con
# estos valores.
# Las variables que tendremos que trabajar son:
# - Date. Representa el dia en el que se recogieron los datos registrados. Para poder trabajar con estos valores, se crearan 3 nuevas columnas, y cada una contendra una 'parte' del dia registrado
#         (dia, mes, anio).
# - Seasons. Corresponde al nombre de la estacion en la que se ha llevado a cabo la recogida de datos. Habra que ver si esta es una potencial covariable predictora del modelo, o habra que eliminarla debido
#            que exista una alta correlacion con respecto a la columna Date. En cualquier caso, por el momento, creare una serie de variables Dummy que sustituiran a esta columna Seasons.
# - Holiday. Cada uno de los ejemplos de esta etiqueta almacena un posible valor binario, pero expresado como cadena de texto. Me asegurare de que unicamente registre dos posibles valores diferentes, 
#            y creare una variable Dummy que represente a esta caracteristica, al igual que con la columna anterior, en codificacion One Hot.
# - Functioning Day. Es mi variable dependiente, y al ser una variable categorica, procesare los registros de la misma llevando a cabo las mismas acciones que con la columna 'Holiday'.
```

#### Columna 'Date'

```python
date_values = dataset['Date'].unique()
date_values[0]
```

```python
# Defino una funcion que me ayude a procesar los valores contenidos en la columna Date

def set_date_ordinal_num(dataset):
    
    """
        Summary:
            Funcion auxiliar encargada de generar un valor numerico para cada uno de los diferentes valores registrados en la columna 'Date' del conjunto de datos.
            Los valores que se asignaran a cada registro corresponde al dia del propio anio, en sentido ordinal.
            
            e.g.: 1/01 ==> 1.
                  2/01 ==> 2.
                  3/01 ==> 3.
        Args:
        
        Returns:
            - date_df ==> DataFrame que registra cada fecha unica del DataFrame del conjunto de datos, indicando cada una de las partes en cuestion de cada dia en una una columna
                          especifica.
        
    """
    date_values = dataset['Date'].unique()
    date_df = pd.DataFrame(columns = ["Date", "Day", "Month", "Year"])
    
    for _ in date_values:
        part_of_day = _.split("/")
        new_row = pd.DataFrame([{"Date": _, "Day": float(part_of_day[0]), "Month": float(part_of_day[1]), "Year": float(part_of_day[2])}])
        date_df = pd.concat([date_df, new_row], ignore_index=True)
        
            
    return date_df
        
    
date_df = set_date_ordinal_num(dataset)    
    
    # Defino el diccionario que mapeara cada uno de los posibles valores
```

```python
date_df
```

```python
# Inserto este nuevo dataframe dentro del dataframe 'dataset' que contiene mi conjunto de datos.

dataset = pd.merge(date_df, dataset, on="Date")  # Indico 'date_df' primero en la lista porque quiero que se situe a la izquierda del DataFrame
dataset
```

```python
# Elimino la columna sobrante 'Date'
dataset.drop(columns = ["Date"], inplace = True)
```

#### Columna 'Seasons'

```python
# A continuacion, genero variables Dummy para la columna Seasons:
unique_values = dataset['Seasons'].unique()
print(unique_values)
```

```python
seasons_df = pd.get_dummies(dataset['Seasons']).astype(int)
seasons_df
```

```python
# Podriamos pensar en incluir este dataframe tal cual esta en en dataframe del conjunto de datos, eliminando la columna real. Esta serie una practica deficiente, y que contribuiria
# a la multicolinealidad de nuestro conjunto de datos, pues existirian variables cuyo valor se podria sumir en funcion del valor del resto de las variables.
# Esto se puede evitar si, tras generar nuevas variables dummy, tenemos en cuenta lo siguiente:   n_variables_dummy = n_posibles_clases - 1.
```

```python
# Elimino una columna de esta dataframe
seasons_df.drop(columns = ['Winter'], inplace = True)
```

```python
# Inserto el dataframe de las estaciones al original con el conjunto de datos
dataset = pd.concat([dataset, seasons_df], axis = 1)

# Elimino la columna Seasons
dataset.drop(columns = ['Seasons'], inplace = True)
```

#### Columna 'Holiday'

```python
# Obtengo los posibles valores que puede contener la columna 'Holiday'
holiday_values = dataset['Holiday'].unique()
holiday_values
```

```python
# Procesar esta columna va a ser bastante sencillo, pues podemos ver como unicamente registra dos valores posibles. En este caso, la variable es binaria, por lo que genero una
# variable Dummy a partir de ella y elimino la original, de forma que pueda mantener la misma informacion, pero en codificacion One Hot.
holiday_values_oh = pd.get_dummies(dataset['Holiday']).astype(int)
holiday_values_oh
```

```python
# Elimino la columna original en el dataframe del conjunto de datos
dataset.drop(columns = ['Holiday'], inplace = True)


# Concateno el dataframe con las variables Dummy al dataframe del conjunto de datos
dataset = pd.concat([dataset, holiday_values_oh], axis = 1)

# Elimino la variable "No Holiday", pues el valor de esta columna se presupone con el valor que registra la nueva columna 'Holiday' (asi, reducimos la posibilidad de multicolinealidad)
dataset.drop(columns = ['No Holiday'], inplace = True)

dataset
```

#### Columna 'Functioning Day'

```python
# Esta columna del dataframe es mi variable dependiente, por lo que no estara incluida en mi matriz de caracteristicas. Sin embargo, la voy a dejar ya procesada, asignandole una codificacion
# One Hot, pues, de igual manera, el modelo no podra entregarme un output que no sea numerico.

funct_days_values = dataset['Functioning Day'].unique()
funct_days_values
```

```python
funct_days_values_oh = pd.get_dummies(dataset['Functioning Day']).astype(int)
funct_days_values_oh
```

```python
# Elimino la columna original en el dataframe del conjunto de datos
dataset.drop(columns = ['Functioning Day'], inplace = True)


# Concateno el dataframe con las variables Dummy al dataframe del conjunto de datos
dataset = pd.concat([dataset, funct_days_values_oh], axis = 1)

# Elimino la variable "Functioning Day", pues el valor de esta columna se presupone con el valor que registra la nueva columna 'Functioning Day' (asi, reducimos la posibilidad de multicolinealidad)
dataset.drop(columns = ['Yes'], inplace = True)


# Renombro la columna recien introducida
dataset = dataset.rename(columns = {"No":"Functioning Day"})

dataset
```
