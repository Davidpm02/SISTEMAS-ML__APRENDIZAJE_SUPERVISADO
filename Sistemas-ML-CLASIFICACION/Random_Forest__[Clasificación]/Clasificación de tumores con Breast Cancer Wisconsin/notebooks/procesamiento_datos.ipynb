# Aclaracion

El dataset actual trata de manera exclusiva el procesamiento y normalizacion del conjunto de datos del que se dispone, a fin de preparar el material para el entrenamiento del modelo de clasificacion

# Importar librerias a utilizar


```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
```

```python
route_dataset = '../dataset/procesed_data.csv'
dataset_df = pd.read_csv(route_dataset)

dataset_df.head(10)
```

```python
# Muestro el tipo de dato de cada una de las columnas del DataFrame
dataset_df.dtypes
```

```python
# Es conveniente que los datos que reciba el modelo compartan todos el mismo tipo de dato (independientemente de que sea numerico, lo cual es imprescindible).
# Por otro lado, se puede ver como algunas columnas contiene valores muy pequenos, en escala [0, 1), mientras que otras contienen valores mucho mas distantes. Es vital aplicar normalizacion
# a todas las columnas del dataset, a fin de entregar unos datos mucho mas uniformes a nuestro modelo.
```

```python
# Hay dos columnas que contienen valores enteros.
# La columna 'id' va a ser descartada, pues no simplemente se utiliza para indexar cada ejemplo registrado en el dataset.
# La columna 'diagnosis', mi etiqueta, almacena valores enteros. Voy a procesar la columna para convertir los valores a flotantes
dataset_df['diagnosis'] = dataset_df['diagnosis'].astype(float)
```

```python
# Reviso que se haya aplicado el cambio
dataset_df.dtypes
```

```python
# Defino mi matriz de caracteristicas y mi matriz de etiquetas
X = dataset_df.drop(columns = ['id', 'diagnosis'])
y = dataset_df['diagnosis']
```

```python
X
```

```python
y
```

```python
# Defino una lista con los nombres de las columna de mi matriz de caracteristicas
x_columns = X.columns
```

```python
# Instancio un objeto de la clase StandardScaler

scaler = StandardScaler()

# Normalizo los valores de mi matriz X
X = scaler.fit_transform(X)
```

```python
X
```

```python
# Defino un nuevo dataframe con los valores normalizados.
# Este dataframe lo usare, principalmente, para poder guardar los datos de manera estructurada, y poder usarlos en el notebook de desarrollo del modelo


norm_dataset = pd.DataFrame(data = X,
                            columns = x_columns)
norm_dataset['diagnosis'] = y
```

```python
norm_dataset
```

```python
# Guardo el dataframe como fichero .csv
import os

norm_dataset__route = '../dataset/'
norm_dataset.to_csv(os.path.join(norm_dataset__route, 'ready_dataset.csv'), index = False)
```
