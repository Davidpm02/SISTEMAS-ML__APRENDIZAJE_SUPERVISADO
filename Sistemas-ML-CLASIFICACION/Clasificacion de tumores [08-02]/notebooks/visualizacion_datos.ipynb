# Analisis objetivos

El objetivo principal de este notebook es el de visualizar la distribucion de datos, y representar de manera grafica informacion del conjunto de datos, que pueda facilitarnos la comprension del mismo.

# Importar librerias

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
route_dataset = '../dataset/procesed_data.csv'
dataset_df = pd.read_csv(route_dataset)

dataset_df.head(10)
```

```python
# Muestro el cuadro de resumen del conjunto de datos en el DataFrame
dataset_df.describe()
```

## Coeficientes de correlacion entre las variables

```python
correlation_matrix = dataset_df.corr()

plt.figure(figsize = (30, 18))
sns.heatmap(correlation_matrix, annot = True)
plt.show()
```

```python
# Podemos ver como contamos con variables que, en general, tienen una muy buena realacion lineal con respecto a nuestra variable dependiente (etiqueta que clasifica al tumor como Maligno o Benigno).
# No obstante, la relacion entre muchas de estas variables con el resto de covariables, tambien es generalmente alta y, si incluyo todas las varialbes en mi matriz de caracteristicas sin hacer nada,
# el modelo seguramente sufra de multicolinealidad.
# Es importante incluir como variables predictoras a caracteristicas que tengan una alta relacion con la variable dependiente, pero que no comparta una gran relacion con el resto de variable de la matriz
# de caracteristicas definida.
# Puedo escoger de forma manual el conjunto de variables predictoras, o puedo hacer uso de regularizacion LASSO/Ridge en mi entrenamiento, para que el modelo comprenda de manera iterativa que variables
# son aquellas que aportan un mayor valor.
```

```python
# Voy a analizar que conjunto de variables de mi dataset tienen un coeficiente de correlacion elevado, a fin de evitar incluir dicho par en mi matriz de caracteristicas.
# Establezco un umbral para la correlación alta
umbral_correlacion_alta = 0.5

# Encuentro pares de correlación alta, excluyendo la diagonal principal
pares_correlacion_alta = [(i, j) for i in range(correlation_matrix.shape[0]) for j in range(i+1, correlation_matrix.shape[1]) if abs(correlation_matrix.iloc[i, j]) > umbral_correlacion_alta]

# Traduzco los índices a nombres de variables
nombres_variables = correlation_matrix.columns
pares_variables_alta_correlacion = [(nombres_variables[i], nombres_variables[j]) for i, j in pares_correlacion_alta]

# Imprimo los pares de variables con alta correlación
for var1, var2 in pares_variables_alta_correlacion:
    print(f"Alta correlación entre: {var1} y {var2} - Correlación: {round(correlation_matrix.loc[var1, var2], 3)}")
```

```python
# Se puede apreciar como la mayoria de columnas del dataset tienen un elevado nivel de correlacion entre si. Probablemente haga uso de regularizacion L2 (Ridge), puesto que considero que, con fines medicos,
# todas las variables del conjunto de datos tienen importante. Este factor de regularizacion me permitira disminuir, hasta valores cercanos a 0, los coeficientes de algunas variables, pero sin llegar a 
# anularlos.
```

```python
# Represento las diferentes distribuciones de ejemplos en las clases posibles de la etiqueta, en funcion de los valores de diferentes columnas
sns.scatterplot(data = dataset_df, x = 'radius_mean', y = 'perimeter_mean', hue = 'diagnosis')
plt.title('Distribución de ejemplos por Perimetro-Radio medio')
plt.xlabel('Radio medio')
plt.ylabel('Perimetro medio')
plt.legend(title='Clase de tumor', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
sns.scatterplot(data = dataset_df, x = 'radius_mean', y = 'smoothness_mean', hue = 'diagnosis')
plt.title('Distribución de ejemplos por Rugosidad-Radio medio')
plt.xlabel('Radio medio')
plt.ylabel('Rugosidad media')
plt.legend(title='Clase de tumor', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
sns.scatterplot(data = dataset_df, x = 'texture_mean', y = 'smoothness_mean', hue = 'diagnosis')
plt.title('Distribución de ejemplos por Textura-Radio medio')
plt.xlabel('Radio medio')
plt.ylabel('Rugosidad media')
plt.legend(title='Clase de tumor', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

En las tres ultimas representaciones se puede apreciar algo de tendencia de existencia de un tumor maligno, en relacion con el valor de radio medio para el nucleo celular del ejemplo registrado.
Parece que, a mayor valor radio del nucleo celular, mayor es el numero de de ejemplos cuyo tumor detectado es maligno
