# Objetivo del notebook

El objetivo principal de este notebook no es otro que el de procesar los datos, y preparar el conjunto de datos para el desarrollo del modelo de clasificacion.

# Importar las librerias a utilizar

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
```

```python
# Cargo el conjunto de datos en memoria como un objeto DataFrame
dataset_route = '../data/hotel_bookings_clean.csv'
bookings_df = pd.read_csv(dataset_route)

# Muestro los 10 primeros registros del dataset
bookings_df.head(10)
```

```python
# Defino la matriz de caracteristicas y matriz de etiquetas
X = bookings_df.drop(columns = ['is_canceled'])
column_names = X.columns

y = bookings_df['is_canceled']
```

```python
# Instancio un objeto de la clase StandardScaler
scaler = StandardScaler()

# Defino un nuevo conjunto  de caracteristicas, con valores normalizados
X = scaler.fit_transform(X)
```

```python
# Muestro los 5 primeros registros de mi matriz de caracteristicas, ahora normalizadas
X[:5, :]
```

```python
# Defino un nuevo DataFrame que contenga los registros normalizados

norm_booking_df = pd.DataFrame(data = X,
                               columns = column_names)

# Inserto la etiqueta al dataframe
norm_booking_df['is_canceled'] = y
```

```python
# Elimino todos los registros nulos del conjunto de datos
norm_booking_df.dropna(inplace = True)
```

```python
# Guardo el dataset normalizado como archivo .csv
import os

norm_dataset__route = "../data/processed/"
if not os.path.exists(norm_dataset__route):
    os.mkdir(norm_dataset__route)
    
norm_booking_df.to_csv(os.path.join(norm_dataset__route, "normalized_hotel_bookings.csv"), index = False)
print("Dataset guardado con exito!")
```
