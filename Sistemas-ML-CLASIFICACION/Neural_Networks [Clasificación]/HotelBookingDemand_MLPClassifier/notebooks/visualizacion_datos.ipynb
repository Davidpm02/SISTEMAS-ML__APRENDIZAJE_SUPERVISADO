# Objetivo del notebook

El objetivo principal de este notebook no es otro que el de visualizar los datos de nuestro dataset.

La visualizacion del dataset permitira obtener informacion sobre tendencias y distribuciones que a priori parecen ocultas, y que nos seran muy utiles al momento de tratar de modelizar la relacion entre nuestra variable dependiente, 'is_canceled', para determinar si, en base a un historico de reservas, una determinada cita va a ser potencialmente cancelada.

# Importar las librerias a utilizar

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargo el conjunto de datos en memoria como un objeto DataFrame
dataset_route = '../data/hotel_bookings_clean.csv'
bookings_df = pd.read_csv(dataset_route)

# Muestro los 10 primeros registros del dataset
bookings_df.head(10)
```

## Distribuciones de datos

```python
### Histograma de visitas canceladas vs activas
plt.figure(figsize = (12, 8))
plt.hist(bookings_df['is_canceled'], color = 'yellowgreen')

plt.title("Distribucion del estado de las reservas")
plt.ylabel('Conteo reservas')
plt.xlabel('Activa  / Cancelada')
plt.xticks(range(0, 2, 1))
plt.show()
```

SÃ­ que es cierto que podemos ver una elevada cifra de reservas que se han cancelado en los Ãºltimos aÃ±os. Puede que se deba a causas razonables, pero viendo los datos, no es algo que podamos estimar de manera sencilla.

```python
## Porcentaje de distribuciones
print('Cantidad de reservas que han sido canceladas:', round(bookings_df[bookings_df['is_canceled'] == 1].shape[0] / bookings_df.shape[0] * 100, 2), '%.')
print('Cantidad de reservas que no se han cancelado:', round(bookings_df[bookings_df['is_canceled'] == 0].shape[0] / bookings_df.shape[0] * 100, 2), '%.')
```

```python
print(bookings_df['arrival_date_month'].unique())
```

```python
# Primero, realizo la agregaciÃ³n por mes
reservas_por_mes = bookings_df.groupby('arrival_date_month').size()

# Ajusto la lista de orden de meses al formato correcto
orden_meses = bookings_df['arrival_date_month'].unique()
reservas_por_mes = reservas_por_mes.reindex(orden_meses)

# Genero un grÃ¡fico de barras con los conteos de cancelaciones
plt.figure(figsize=(15, 8))
plt.bar(reservas_por_mes.index, reservas_por_mes.values, color='yellowgreen')

plt.title('Histograma de reservas emitidas por meses')
plt.ylabel('Conteo de reservas')
plt.xlabel('Meses registrados')
plt.xticks(range(1, 13, 1))
plt.show()
```

```python
# Primero, realizo la agregaciÃ³n por mes contando las cancelaciones
cancelaciones_por_mes = bookings_df[bookings_df['is_canceled'] == 1].groupby('arrival_date_month').size()

# Ajusto la lista de orden de meses al formato correcto
orden_meses = bookings_df['arrival_date_month'].unique()
cancelaciones_por_mes = cancelaciones_por_mes.reindex(orden_meses)

# Genero un grÃ¡fico de barras con los conteos de cancelaciones
plt.figure(figsize=(15, 8))
plt.bar(cancelaciones_por_mes.index, cancelaciones_por_mes.values, color='red')

plt.title('Histograma de reservas canceladas por meses')
plt.ylabel('Conteo de cancelaciones')
plt.xlabel('Meses registrados')
plt.xticks(range(1, 13, 1))
plt.show()
```

```python
## Grafico de barras que enfrenta el registro de reservas vs el conteo de cancelaciones para cada mes del que se tiene registro

# ParÃ¡metros para el grÃ¡fico
bar_width = 0.35 
index = np.arange(len(reservas_por_mes))  

# CreaciÃ³n del grÃ¡fico
plt.figure(figsize=(15, 8))

# Barras de reservas
plt.bar(index - bar_width/2, reservas_por_mes.values, bar_width, color='yellowgreen', label='Reservas')

# Barras de cancelaciones
plt.bar(index + bar_width/2, cancelaciones_por_mes.values, bar_width, color='red', label='Cancelaciones')

plt.title('Histograma de reservas emitidas por meses')
plt.ylabel('Conteo')
plt.xlabel('Meses registrados')
plt.xticks(index, reservas_por_mes.index)  # Asegurando que las etiquetas del eje X sean los meses
plt.legend()  # AÃ±ade una leyenda para distinguir reservas de cancelaciones

plt.show()
```

```python
## Conteo de reservas, canceladas o no , segun el tamano de la familia

# Para esto, defino 4 grupos, con los siguientes integrantes:
#  - 1 persona.
#  - 2 personas.
#  - 3 personas.
#  - +3 personas.

# Para esto, voy a tener en cuenta los valores de las columnas 'adults', 'children', 'babies' para cada ejemplo, y filtrare si el registro en cuestion es de una 
# reserva que ha sido cancelada o no

booking_1person = 0
booking_2person = 0
booking_3person = 0
booking_more3person = 0

for resultado in bookings_df.iterrows():
    
    family_members = resultado[1][7] + resultado[1][8] + resultado[1][9]

    if family_members == 1.0:
        booking_1person += 1
    elif family_members == 2.0:
        booking_2person += 1
    elif family_members == 3.0:
        booking_3person += 1
    elif family_members > 3.0:
        booking_more3person += 1
```

```python
# Muestro en un grafico de barras la distribucion de cancelaciones en funcion del tamanio de la familia
bookings = {
    "1 persona":booking_1person,
    "2 personas": booking_2person,
    "3 personas": booking_3person,
    "+3 personas": booking_more3person
}

# Representacion
plt.figure(figsize = (12, 8))
for key, value in bookings.items():
    plt.bar(key, value, color="yellowgreen")

plt.title("DistribuciÃ³n de reservas por tamaÃ±o de familias")
plt.ylabel("Conteo de reservas")
plt.xlabel("Integrantes familias")

plt.xticks(list(bookings.keys()))
plt.grid(True)
plt.show()
```

```python
## Conteo de cancelaciones segun el tamano de la familia.

# Para esto, defino 4 grupos, con los siguientes integrantes:
#  - 1 persona.
#  - 2 personas.
#  - 3 personas.
#  - +3 personas.

# Para esto, voy a tener en cuenta los valores de las columnas 'adults', 'children', 'babies' para cada ejemplo, y filtrare si el registro en cuestion es de una 
# reserva que ha sido cancelada o no

canceled_booking_1person = 0
canceled_booking_2person = 0
canceled_booking_3person = 0
canceled_booking_more3person = 0

for resultado in bookings_df.iterrows():
    if resultado[1][0] == 1.0:  # Si la reserva ha sido cancelada...
        family_members = resultado[1][7] + resultado[1][8] + resultado[1][9]

        if family_members == 1.0:
            canceled_booking_1person += 1
        elif family_members == 2.0:
            canceled_booking_2person += 1
        elif family_members == 3.0:
            canceled_booking_3person += 1
        elif family_members > 3.0:
            canceled_booking_more3person += 1
```

```python
# Muestro en un grafico de barras la distribucion de cancelaciones en funcion del tamanio de la familia
canceled_bookings = {
    "1 persona":canceled_booking_1person,
    "2 personas": canceled_booking_2person,
    "3 personas": canceled_booking_3person,
    "+3 personas": canceled_booking_more3person
}

# Representacion
plt.figure(figsize = (12, 8))
for key, value in canceled_bookings.items():
    plt.bar(key, value, color="yellowgreen")

plt.title("DistribuciÃ³n de cancelaciones por tamaÃ±o de familias")
plt.ylabel("Conteo de cancelaciones")
plt.xlabel("Integrantes familias")

plt.xticks(list(canceled_bookings.keys()))
plt.grid(True)
plt.show()
```

```python
## Sabiendo el numero de reservas emitidas de cada familia, en funcion de su tamanio, y el numero de cancelaciones para cada tipo de familia, puedo obtener
## el porcentaje de reservas canceladas para cada caso

# 1 integrante
print("De las reservas emitidas por familias de 1 solo integrante, en los ultimos 2 anios han sido anuladas el {} % de las mismas.".format(round(canceled_booking_1person / booking_1person * 100, 2)), end = "\n")
print()
# 2 integrantes
print("De las reservas emitidas por familias de 2 integrantes, en los ultimos 2 anios han sido anuladas el {} % de las mismas.".format(round(canceled_booking_2person / booking_2person * 100, 2)), end = "\n")
print()
# 3 integrantes
print("De las reservas emitidas por familias de 3 integrantes, en los ultimos 2 anios han sido anuladas el {} % de las mismas.".format(round(canceled_booking_3person / booking_3person * 100, 2)), end = "\n")
print()
# +3 integrantes
print("De las reservas emitidas por familias de mas de 3 integrantes, en los ultimos 2 anios han sido anuladas el {} % de las mismas.".format(round(canceled_booking_more3person / booking_more3person * 100, 2)), end = "\n")
print()
```

Podemos ver entonces como existe una tendencia bastante notable a cancelar las reservas del hotel para familias de mas de 3 personas, habiendose registrado un **41.51%** de cancelaciones en este caso.

Para el resto de tamanios de las familias, se aprecia como, en familias de 2 integrantes, el ratio de cancelacion de reservas es tambien superior al resto de los casos, con cerca de un **40%** de cancelaciones.

Si bien el ratio de cancelacion es similar en ambos casos, el numero de cancelaciones para habitaciones de 2 integrantes es muy superior, debido al superior conteo original de reservas para este tipo de familias.

```python
## Tendencia de cancelacion de las visitas en funcion del numero de dias entre el dia de la reserva, y el dia de llegada al hotel

# Grafico primero el numero de reservas para cada uno de los siguientes grupos
# Igual que antes, defino una serie de grupos para cada tipo de reserva
booking_in20days = bookings_df[bookings_df['lead_time'] <= 20].count().values[0]
booking_in60days = bookings_df[(bookings_df['lead_time'] > 20) & (bookings_df['lead_time'] <= 60)].count().values[0]
booking_in100days = bookings_df[(bookings_df['lead_time'] > 60) & (bookings_df['lead_time'] <= 100)].count().values[0]
booking_in150days = bookings_df[(bookings_df['lead_time'] > 100) & (bookings_df['lead_time'] <= 150)].count().values[0]
booking_in200days = bookings_df[(bookings_df['lead_time'] > 150) & (bookings_df['lead_time'] <= 200)].count().values[0]
booking_in300days = bookings_df[(bookings_df['lead_time'] > 200) & (bookings_df['lead_time'] <= 300)].count().values[0]
booking_inMoreThan300days = bookings_df[bookings_df['lead_time'] < 300].count().values[0]
```

```python
# Muestro en un grafico de barras la distribucion de reservas en funcion del numero de dias entre el dia de la reserva, y el dia de llegada al hotel
bookings = {
    "20 dias o menos":booking_in20days,
    "Entre 21 y 60 dias": booking_in60days,
    "Entre 61 y 100 dias": booking_in100days,
    "Entre 101 y 150 dias": booking_in150days,
    "Entre 151 y 200 dias": booking_in200days,
    "Entre 201 y 300 dias": booking_in300days,
    "Mas de 301 dias": booking_inMoreThan300days
}

# Representacion
plt.figure(figsize = (12, 8))
for key, value in bookings.items():
    plt.bar(key, value, color="yellowgreen")

plt.title("DistribuciÃ³n de reservas por dias de transicion")
plt.ylabel("Conteo de reservas")
plt.xlabel("Intervalo de dias de transicion")

plt.xticks(list(bookings.keys()), rotation = 35)
plt.grid(True)
plt.show()
```

```python
# Muestro en un grafico de barras la distribucion de cancelaciones en funcion del numero de dias entre el dia de la reserva, y el dia de llegada al hotel
canceled_bookings = {
    "20 dias o menos":canceled_booking_in20days,
    "Entre 21 y 60 dias": canceled_booking_in60days,
    "Entre 61 y 100 dias": canceled_booking_in100days,
    "Entre 101 y 150 dias": canceled_booking_in150days,
    "Entre 151 y 200 dias": canceled_booking_in200days,
    "Entre 201 y 300 dias": canceled_booking_in300days,
    "Mas de 301 dias": canceled_booking_inMoreThan300days
}

# Representacion
plt.figure(figsize = (12, 8))
for key, value in canceled_bookings.items():
    plt.bar(key, value, color="red")

plt.title("DistribuciÃ³n de cancelaciones por dias de transicion")
plt.ylabel("Conteo de cancelaciones")
plt.xlabel("Intervalo de dias de transicion")

plt.xticks(list(canceled_bookings.keys()), rotation = 35)
plt.grid(True)
plt.show()
```
